```python
import pandas as pd

salary_df = pd.read_csv('../../data/Data Science Job Salary dataset/datascience_salaries.csv')

salary_df.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

salary_df['salary_normalized'] = scaler.fit_transform(salary_df[['salary']])

salary_df.head()
```

```python
from sklearn.decomposition import PCA

numeric_cols = salary_df.select_dtypes(include=['float64', 'int64']).columns
numeric_data = salary_df[numeric_cols]

pca = PCA(n_components=2)
principal_components = pca.fit_transform(numeric_data)

salary_df['PC1'] = principal_components[:, 0]
salary_df['PC2'] = principal_components[:, 1]

salary_df.head()
```

```python
grouped_salary = salary_df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()

print(grouped_salary)
```
