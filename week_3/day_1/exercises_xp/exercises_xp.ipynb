```python
iris_df = pd.read_csv('../data/Iris Dataset/Iris.csv')

iris_df.info()

# RangeIndex: 150 entries, 0 to 149
# Data columns (total 6 columns):
#  #   Column         Non-Null Count  Dtype    Category
# ---  ------         --------------  -----    --------
#  0   Id             150 non-null    int64    Quant : number
#  1   SepalLengthCm  150 non-null    float64  Quant : is a number
#  2   SepalWidthCm   150 non-null    float64  Quant : is a number
#  3   PetalLengthCm  150 non-null    float64  Quant : is a number
#  4   PetalWidthCm   150 non-null    float64  Quant : is a number
#  5   Species        150 non-null    object   Qaul : is a category
```

```python
mean_sepal_length_cm = iris_df['SepalLengthCm'].mean()
median_sepal_length_cm = iris_df['SepalLengthCm'].median()
mode_sepal_length_cm = iris_df['SepalLengthCm'].mode()[0]

print('Mean:', mean_sepal_length_cm)
print('Median:', median_sepal_length_cm)
print('Mode:', mode_sepal_length_cm)

import matplotlib.pyplot as plt
plt.plot(iris_df['SepalLengthCm'])
plt.show()
```
