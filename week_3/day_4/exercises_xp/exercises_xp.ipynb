```python
# ex 4

import pandas as pd
import seaborn as sns

df = pd.read_csv("../../data/Student Mental health/Student Mental health.csv")
```

```python
def cgpa_to_numeric(range_str):
    low, high = range_str.split(" - ")
    return (float(low) + float(high)) / 2

df['CGPA_numeric'] = df['What is your CGPA?'].apply(cgpa_to_numeric)
```

```python
sns.histplot(df['CGPA_numeric'], bins=10, color='skyblue', kde=False)
plt.xlabel("CGPA")
plt.ylabel("Number of Students")
plt.title("Distribution of Students' CGPA")
plt.show()
```

```python
# ex 5

anxiety_counts = df.groupby(['Choose your gender', 'Do you have Anxiety?']).size().reset_index(name='count')

sns.barplot(data=anxiety_counts, x='Choose your gender', y='count', hue='Do you have Anxiety?', palette='pastel')
plt.xlabel("Gender")
plt.ylabel("Number of Students")
plt.title("Anxiety Levels Across Genders")
plt.show()
```

```python
# ex 6

df['Panic_numeric'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

sns.scatterplot(data=df, x='Age', y='Panic_numeric', hue='Panic_numeric', palette={0:'blue', 1:'red'}, alpha=0.6)
plt.xlabel("Age")
plt.ylabel("Panic Attack (0=No, 1=Yes)")
plt.title("Relationship Between Age and Panic Attacks")
plt.yticks([0,1], ['No', 'Yes'])
plt.show()
```
